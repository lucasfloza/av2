```python
import mermaid as md

render = md.Mermaid("""
graph TD
    Title["Arquivo: monte-carlo-pi.py"]
    
    Title --> A[Inicio do programa]
    A --> B[Define num_iterations = 1000]
    B --> C[Chama estimate_pi com num_iterations]
    
    %% Função estimate_pi
    C --> D[Inicializa num_points_inside_circle = 0]
    D --> E[Inicializa total_num_points = 0]
    E --> F[Inicializa pi_estimates = lista vazia]
    
    F --> G[Loop: i de 0 a num_iterations]
    G --> H[Gera x aleatorio entre 0 e 1]
    H --> I[Gera y aleatorio entre 0 e 1]
    I --> J[Calcula distance = x² + y²]
    
    J --> K{distance <= 1?}
    K -->|Sim| L[Incrementa num_points_inside_circle]
    K -->|Não| M[Não incrementa]
    L --> N[Incrementa total_num_points]
    M --> N
    
    N --> O[Calcula pi_estimate = 4 * pontos_dentro / total_pontos]
    O --> P[Adiciona pi_estimate à lista pi_estimates]
    P --> Q[Plota ponto no grafico: i+1 vs pi_estimate]
    
    Q --> R{i < num_iterations-1?}
    R -->|Sim| G
    R -->|Não| S[Retorna pi_estimate final]
    
    %% Volta ao programa principal
    S --> T[Adiciona linha horizontal com valor real de π]
    T --> U[Configura rotulos dos eixos]
    U --> V[Configura titulo do grafico]
    V --> W[Adiciona legenda e grade]
    W --> X[Exibe o grafico]
    X --> Y[Fim do programa]
""")

render
```

## C)

O código estima o valor de π usando o método de Monte Carlo. Ele gera pontos aleatórios dentro de um quadrado de lado 1 e conta quantos caem dentro de um quarto de círculo inscrito nesse quadrado. A proporção de pontos dentro do círculo em relação ao total se aproxima de π/4, e multiplicando essa razão por 4, obtém-se uma estimativa de π.

## D)

Quanto maior o número de tentativas (N), mais próxima a estimativa de π fica do valor real. Para valores pequenos de N, as flutuações são grandes, e o valor de π varia bastante. Com N grande, a curva se estabiliza e converge lentamente para π ≈ 3.14159, pois o efeito do acaso vai sendo compensado pela grande quantidade de amostras. Isso mostra que o método de Monte Carlo é confiável, mas exige muitas iterações para alcançar boa precisão.
