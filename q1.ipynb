```python
def is_prime(num):
    """Check if a number is prime"""
    if num < 2:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

def find_primes(quantity):
    """Find a specific quantity of prime numbers"""
    primes = []
    number = 2

    while len(primes) < quantity:
        if is_prime(number):
            primes.append(number)
        number += 1

    return primes

# Find the first 300 prime numbers
first_300_primes = find_primes(300)

# Print the numbers with formatting
print("The first 300 prime numbers are:")
for i, prime in enumerate(first_300_primes, 1):
    print(f"{i:3d}¬∫ prime: {prime:4d}")
```
